### 2. Apa saja jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016?

```python
df2 = client.query('''
SELECT DISTINCT description, year
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016
GROUP BY year , description
''').to_dataframe()

df2
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016 sebanyak 54 jenis kejahatan. Hal ini dapat menjadikan pertimbangan oleh pihak kepolisian

### 3. Berapa banyak kasus tindak kejahatan berdasarkan kasus pembersihannya (clearance status) selama tahun 2016?

```python
df3 = client.query('''
SELECT 	COUNT(clearance_status) AS kasus_tindak_kejahatan
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016
''').to_dataframe()

df3
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa terdapat kasus tindak kejahatan berdasarkan kasus pembersihannya 35742 kasus, hal ini dapat menjadikan bahan pertimbangan pihak kepolisian guna memberikan treatment yang lebih supaya kasus tindakan kejahatan berkurang

### 4. Jenis kejahatan apa saja yang paling banyak belum terselesaikan kasusnya di tahun 2016?

```python
df4 = client.query('''
SELECT year, description,
COUNT(clearance_status) AS total_belum_terselesaikan
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND clearance_status = 'Not cleared'
GROUP BY year, description
ORDER BY total_belum_terselesaikan DESC
''').to_dataframe()

df4
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa jenis kejahatan yang paling banyak belum terselesaikan kasusnya di tahun 2016 adalah jenis kejahatan **THEFT** atau pencurian yaitu sebanyak 8607. Hal ini perlu diperhatikan oleh pihak kepolisisan guna segera dapat terselesaikan

### 5. Di bulan apa di tahun 2016 kasus pencurian (theft) sangat banyak terjadi?

```python
df5 = client.query('''
SELECT  year, EXTRACT(month FROM timestamp) AS month, description,
COUNT(description) AS total_kasus_pencurian 
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND description = 'THEFT                         '
GROUP BY year,month, description
ORDER BY total_kasus_pencurian DESC
''').to_dataframe()

df5
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa kasus pencurian (theft) yang sangat banyak terjadi di tahun 2016 adalah pada bulan Desember yaitu sebanyak 925. Hal ini bisa saja disebabkan meningkatnya pengunjung di kota Austin yang bertepatan dengan peringataan hari natal atau cuti bersama, sehingga kasus pencurian di kota Austin meningkat. Perihal ini  perlu menjadi perhatian khusus bagi pihak kepolisisan untuk meningkatkan keamaan dibulan - bulan tertentu, guna meminimalisir tinggak kejahatan


```python
df5_1 = client.query('''
SELECT SUM(total_kasus_pencurian) AS jumlah_kasus_pencurian
FROM(SELECT  year, EXTRACT(month FROM timestamp) AS month, description,
COUNT(description) AS total_kasus_pencurian 
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND description = 'THEFT                         '
GROUP BY year,month, description
ORDER BY total_kasus_pencurian DESC)
''').to_dataframe()

df5_1
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa jumlah total kasus pencurian pada tahun 2016 adalah sebanyak 10442 kasus

### 6. Di distrik apa yang paling banyak terjadi kasus pencurian selama 2016?

```python
df6 = client.query('''
SELECT year, district,
COUNT(description) AS total_kasus_pencurian
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND description = 'THEFT                         '
GROUP BY year, district
ORDER BY total_kasus_pencurian DESC
''').to_dataframe()

df6
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa kasus pencurian terbanyak yang terjadi pada tahun 2016 terjadi di distrik D sebanyak 1597 kasus pencurian. Hal ini perlu di perhatikan oleh pihak kepolisian guna meningkatkan keamaan di distrik D, sehingga terjadi penurunan kasus pencurian atau jenis kejahatan lainnya

### 7. Berapa lama rata-rata waktu (dalam hari) kasus tindak kejahatan 'theft' terselesaikan sejak kasus tersebut terangkat selama tahun 2016? (Hint: Hitung selisih hari dari timestamp hingga clearance date)

```python
df6 = client.query('''
SELECT AVG(selisih_hari) AS rata_rata_selisih_hari
FROM(SELECT year, description, timestamp, clearance_date,
DATE_DIFF(clearance_date, timestamp, day) AS selisih_hari
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND description = 'THEFT                         ')
''').to_dataframe()

df6
```

> Berdasarkan hasil output diatas di peroleh informasi bahwa lama rata-rata waktu (dalam hari) kasus tindak kejahatan 'theft' terselesaikan sejak kasus tersebut terangkat selama tahun 2016 adalah 21 hari, yang berarti pihak kepolisian dapat menyelesaikan kasus kejahatan jenis pencurian selama rata-rata 21 hari pada tahun 2016
