Dan terdapat pula informasi missing values di setiap kolom pada data. Dari informasi di atas hampir semua kolom terdapat missing value, dan hanya 3 kolom yang tidak terdapat missing value yaitu kolom country_region, date dan confirmed

> Code di atas merupakan code untuk menampilkan jumlah missing value di setiap kolom yang mana kolom active merupakan kolom terbanyak yang terdapat missing value yaitu sebanyak 50502

```python
df['province_state'].fillna('Other', inplace=True)
df['location_geom'].fillna('Other', inplace=True)
df['deaths'].fillna(0, inplace=True)
df['recovered'].fillna(0, inplace=True)
df['active'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

> Code di atas merupakan code untuk menampilkan jumlah missing value di setiap kolom dan setelah melakukan perubahan value di setiap kolom yang terdapat missing value, sudah tidak ada lagi missing value di setiap kolom

```python
df.info()
```

> Code df.info() untuk menampilkan informasi dan memastikan kembali bahwa di setiap kolom tidak terdapat missing value

## **4.2 Mengganti Nama Kolom**

```python
New1_df = df.copy()
New1_df.head() 
```

> Code df.copy() merupakan code untuk mengcopy data menjadi data baru, supaya apabila melakukan manipulasi terhadap data baru tidak akan berpengaruh terhadap data lama

```python
renamed_df = New1_df.rename(
    columns={
        'province_state': 'provinsi', 
        'country_region': 'negara',
        'date' : 'tanggal',
        'location_geom' : 'letak_geografis',
        'confirmed' : 'dikonfirmasi',
        'deaths' : 'meninggal',
        'recovered' : 'sembuh',
        'active' : 'aktif'
        }
)

renamed_df.head()
```

> Code di atas merupakan untuk merubah nama kolom pada data, yang mana sebelumnya nama kolom menggunakan bahasa inggris di rubah menjadi bahasa indonesa

## **4.3 Menambahkan Kolom**

```python
New2_df = renamed_df.copy()
```

```python
New2_df['persentase_meninggal(%)'] = (New2_df.meninggal / New2_df.dikonfirmasi)*100
New2_df.shape
```

> Code di atas merupakan code untuk menambahkan kolom baru, yang mana kolom baru ini berupa kolom persentase yang meninggal dari data kolom dikonfirmasi. sehingga dengan menggunakan .shape kita bisa mengetahui jumlah baris dan kolom, dan terdapat penambahan kolom yang awalnya 8 menjadi 9 kolom

```python
pd.set_option("display.precision", 2)
New2_df
```

## **4.4 Menghapus Kolom**

```python
New2_df.drop(['letak_geografis'], inplace = True, axis = 1)
```

```python
New2_df.shape
```

> Code di atas merupakan code untuk menghapus kolom, dimana kida dapat menghapus kolom yang dirasa tidak digunakan dalam analisis. pada contoh ini kolom yang dihapus adalah kolom letak geografis dengan menggunakan .drop()

## **5.2 Grouping and Aggregating**

```python
New2_df
```

> Pada query data, data yang digunakan adalah data yang sudah dimanipulasi

```python
Negara_China_1 = New2_df[
    (New2_df['negara'] == 'China') 
].groupby(['negara', 'provinsi'])['meninggal'].max().sort_values(ascending=False).reset_index()
Negara_China_1
```

> Code di atas merupakan code untuk grouping kolom pada data dengan menampilkan nilai maksimum kematian pada setiap valus kolom provinsi dan negara. Pada code di atas kita menentukan grouping untuk negara china

```python
Negara_China_2 = New2_df[
    (New2_df['negara'] == 'China') &
    (New2_df['tanggal'].str.startswith('2021'))
].groupby(['negara', 'provinsi'])['meninggal'].mean().sort_values(ascending=False).reset_index()
Negara_China_2
```

> Code di atas merupakan code untuk grouping kolom pada data dengan kondisi data pada tahun 2021 serta menampilkan rata - rata kematian pada setiap valus kolom provinsi dan negara. Pada code di atas kita menentukan grouping untuk negara china pada tahun 2021

```python
Negara_China_3 = New2_df[
    (New2_df['negara'] == 'China') &
    (New2_df['provinsi'] == 'Hubei') &
    (New2_df['tanggal'].str.startswith('2020'))
].groupby(['negara', 'provinsi','tanggal'])['meninggal'].sum().sort_values(ascending=False).reset_index()
Negara_China_3
```

> Code di atas merupakan code untuk grouping kolom pada data dengan kondisi negara == china, provinsi == hubei dan pada tahun 2020 serta menampilkan total jumlah kematian pada setiap valus kolom provinsi, negara dan tanggal. Pada code di atas kita menentukan grouping untuk negara china, provinsi hubei dan tanggal pada tahun 2020

```python
Negara_China_4 = New2_df[
    (New2_df['negara'] == 'China') &
    (New2_df['persentase_meninggal(%)'] == 0.0) 
].groupby(['negara', 'provinsi'])['meninggal'].sum().sort_values(ascending=False).reset_index()
Negara_China_4
```

> Code di atas merupakan code untuk grouping kolom pada data dengan kondisi negara == china dan persentase kematian 0 % serta menampilkan total jumlah kematian pada setiap valus kolom provinsi yang memiliki kasus kematian persentase 0 %. Pada code di atas kita menentukan grouping untuk negara china dengan persentase kematian 0%

```python
Negara_China_5 = New2_df[
    (New2_df['negara'] == 'China') &
    (New2_df['tanggal'].str.startswith('2020'))
].groupby(['negara', 'provinsi']).agg({'dikonfirmasi':'sum','meninggal':'sum'}).reset_index()
Negara_China_5
```

> Code di atas merupakan code untuk grouping kolom pada data dengan kondisi negara == china dan tanggal untuk tahun 2020 serta menampilkan totaol jumlah kasus yang dikonfirmasi dan total jumlah kasus meninggal dengan menggunakan argumen .agg . Pada code di atas kita menentukan grouping untuk negara china pada tahun 2020
